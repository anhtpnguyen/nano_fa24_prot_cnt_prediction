Processing output from PROSE

```python
import h5py
```

```python
data = []
protein_names = []
data_dict = {}
```

```python
fixing_protname = {
    "FAM90A17P OS=Homo sapiens OX=9606 GN=FAM90A16P PE=5 SV=1":"Putative protein FAM90A16P/FAM90A17P",
    "arylesterase 1 OS=Homo sapiens OX=9606 GN=PON1 PE=1 SV=3":"Serum paraoxonase/arylesterase 1",
    "reductase family 9C member 7 OS=Homo sapiens OX=9606 GN=SDR9C7 PE=1 SV=1": "Short-chain dehydrogenase/reductase family 9C member 7",
    "threonine-protein kinase MRCK beta OS=Homo sapiens OX=9606 GN=CDC42BPB PE=1 SV=2" : "Serine/threonine-protein kinase MRCK beta"
}
```

```python
with h5py.File("prose_data/all.h5") as f:
    for item in f.keys():
        #print(len(f[item]))
        if len(f[item]) == 1:
            test = f[item]
            for group in test:
                name_real = fixing_protname[str(group)]
                protein_names.append(name_real)
                embed = test[group]
                print(group)
                data.append(embed[:])
                data_dict[name_real] = embed[:]
        else:
            embed = f[item]
            protname = embed.name
            string_list = protname.split(" ")
            protname_formatted = ""
            notOS = True
            for string in string_list[1:]:
                # while "OS" not in string:
                if "OS" in string:
                    notOS = False
                if notOS:
                    protname_formatted += string + ' '

            name_real = protname_formatted[:-1]
            protein_names.append(name_real)
            data.append(embed[:])
            data_dict[name_real] = embed[:]
        
```
